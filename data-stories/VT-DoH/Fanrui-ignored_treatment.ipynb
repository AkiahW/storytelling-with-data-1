```python
reduction_max= pd.read_csv('https://github.com/FanruiShao/storytelling-with-data/raw/master/data-stories/VT-DoH/loss_spending.csv')
```

```python
reduction_max
```

```python
import plotly.express as px

fig = px.bar(reduction_max, x='specialty', y='actual spending', title='in $millions')
fig.show()
```

```python
import plotly.express as px

fig = px.bar(reduction_max, x='specialty', y='expected spending')
fig.show()
```

```python
df = pd.concat([reduction_max, reduction_max.copy()], axis=0).reset_index()
```

```python
df ['spending in $millions'] = 0
df ['medicare specialty'] = ''
```

```python
df
```

```python
df.loc[0:2]['spending in $millions'] = df.loc[0:2]['expected spending']
df.loc[3:]['spending in $millions'] = df.loc[3:]['actual spending']

df.loc[0:2]['medicare specialty'] = 'expected'
df.loc[3:]['medicare specialty'] = 'actual'

df
```

```python
px.bar(data_frame=df, x= 'specialty', y='spending in $millions',color='medicare specialty', barmode='group')
```

```python
state_drop = pd.read_csv('https://github.com/FanruiShao/storytelling-with-data/raw/master/data-stories/VT-DoH/state_drop.csv')
```

```python
state_drop
```

```python
px.bar(data_frame=state_drop, x= 'state', y='percentage drop(%)', color = 'actual spending ($millions)',  hover_name = 'percentage drop(%)' , title ='cumulative reduction in MPFS Spending, 2020')
```

```python
df = pd.concat([state_drop, state_drop.copy()], axis=0).reset_index()
df ['spending($millions)'] = 0
df ['type'] = ''
df
```

```python
df.loc[0:5]['spending($millions)'] = df.loc[0:5]['expected spending ($millions)']
df.loc[6:]['spending($millions)'] = df.loc[6:]['actual spending ($millions)']

df.loc[0:5]['type'] = 'expected'
df.loc[6:]['type'] = 'actual'

df
```

```python
px.bar(data_frame = df, x = 'state', y = 'spending($millions)',color='type', barmode='group', title = 'MPFS Spending, 2020')
```
